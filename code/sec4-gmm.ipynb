````python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
# Cargar el dataset de CÃ¡ncer de Mama
data = load_breast_cancer(as_frame=True)
X = data.data
````

````python
# plot the breast cancer data
plt.hist(X['mean radius'], bins=30, alpha=0.7, color='blue');
````

Encontrar "subgrupos" usando K-means (no supervisado).

````python
# 1. SELECCIÃN DE CARACTERÃSTICAS PARA VISUALIZACIÃN
# Usaremos 'mean radius' y 'mean texture' para poder graficar en 2D
X_vis = X[['mean radius']]

# 2. ESCALADO DE DATOS
# Es una buena prÃ¡ctica escalar, incluso para una sola dimensiÃ³n.
# Se necesita remodelar los datos para el scaler.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_vis.values.reshape(-1, 1))

# 3. AJUSTE DEL MODELO K-MEANS
# Ajustamos el modelo en los datos escalados de una dimensiÃ³n
kmeans = KMeans(n_clusters=2, n_init='auto', random_state=42)
kmeans.fit(X_scaled)

# Obtenemos las etiquetas y los centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# AÃ±adimos las etiquetas de cluster al DataFrame original para facilitar el filtrado
X_vis['cluster'] = labels

# 4. VISUALIZACIÃN CON HISTOGRAMAS
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 7))

# Datos para cada cluster
cluster0_data = X_vis[X_vis['cluster'] == 0]['mean radius']
cluster1_data = X_vis[X_vis['cluster'] == 1]['mean radius']

# Graficar histogramas superpuestos para cada cluster
# Usamos `alpha` para la transparencia y poder ver ambos
sns.histplot(cluster0_data, ax=ax, color='skyblue',
             label='Cluster 0', kde=True)
sns.histplot(cluster1_data, ax=ax, color='coral', label='Cluster 1', kde=True)

# Revertir la transformaciÃ³n de los centroides para graficarlos en la escala original
centroids_original_scale = scaler.inverse_transform(centroids)

# AÃ±adir lÃ­neas verticales para los centroides
ax.axvline(centroids_original_scale[0][0], color='blue',
           linestyle='--', linewidth=2, label='Centroide 0')
ax.axvline(centroids_original_scale[1][0], color='red',
           linestyle='--', linewidth=2, label='Centroide 1')

ax.set_title('Clustering K-Means en una DimensiÃ³n (Mean Radius)')
ax.set_xlabel('Radio Medio (Escala Original)')
ax.set_ylabel('Frecuencia')
ax.legend()

plt.show()
````

GMM en lugar de K-means.

$$
\begin{aligned}
& \underset{\mu_1, \mu_2, \sigma_1, \sigma_2, \pi_1}{\text{minimizar}}
& & -\sum_{i=1}^{N} \ln \left( \pi_1 \mathcal{N}(x_i | \mu_1, \sigma_1^2) + (1-\pi_1) \mathcal{N}(x_i | \mu_2, \sigma_2^2) \right) \\
& \text{s.a.}
& & 0 \le \pi_1 \le 1 \\
& & & \sigma_1 > 0, \quad \sigma_2 > 0
\end{aligned}
$$

Primero definir la funciÃ³n objetivo y las restricciones.

````python
# Definir la funciÃ³n de verosimilitud negativa para el modelo de mezcla gaussiana (GMM)
def gmm_verosimilitud_negativa(params):
    # Desempaquetar los parÃ¡metros
    mu1, mu2, sigma1, sigma2, pi_1 = params
    # Calcular la proporciÃ³n del segundo componente
    pi_2 = 1 - pi_1

    # Crear las distribuciones normales para cada componente
    rv1 = norm(loc=mu1, scale=sigma1)  # Componente 1
    rv2 = norm(loc=mu2, scale=sigma2)  # Componente 2

    # Calcular la mezcla de las dos distribuciones
    mezcla_rv = pi_1 * rv1.pdf(X_radius) + pi_2 * rv2.pdf(X_radius)
    
    # Calcular la verosimilitud negativa (logaritmo negativo de la probabilidad)
    verosimilitud_negativa = -np.sum(np.log(mezcla_rv))
    return verosimilitud_negativa
````

La funciÃ³n `minimize` de `scipy.optimize` es una herramienta poderosa para resolver problemas de optimizaciÃ³n. Sus argumentos principales son:

1. **fun**: La funciÃ³n objetivo que se desea minimizar. En este caso, serÃ¡ la funciÃ³n de verosimilitud negativa de GMM.

2. **x0**: Un array o lista con los valores iniciales de los parÃ¡metros que se optimizarÃ¡n. En este caso, se usarÃ¡n los valores iniciales de los parÃ¡metros del modelo GMM.

3. **method**: El mÃ©todo de optimizaciÃ³n a utilizar. Algunos mÃ©todos comunes son:
  - `'L-BFGS-B'`: MÃ©todo limitado de BFGS para problemas con restricciones de caja.
  - `'Nelder-Mead'`: MÃ©todo simplex para optimizaciÃ³n sin restricciones.
  - `'TNC'`: MÃ©todo de Newton truncado para problemas con restricciones.

4. **bounds**: Una lista de tuplas que define los lÃ­mites inferiores y superiores de cada parÃ¡metro. Es Ãºtil para garantizar que los parÃ¡metros se mantengan dentro de un rango vÃ¡lido.

5. **constraints**: Una lista de restricciones adicionales que deben cumplirse durante la optimizaciÃ³n. Estas pueden ser de igualdad o desigualdad.

6. **args**: Una tupla con argumentos adicionales que se pasarÃ¡n a la funciÃ³n objetivo. Esto es Ãºtil si la funciÃ³n requiere datos adicionales.

7. **options**: Un diccionario para configurar opciones especÃ­ficas del mÃ©todo, como el nÃºmero mÃ¡ximo de iteraciones (`maxiter`) o la tolerancia (`tol`).

Ejemplo de uso para minimizar la funciÃ³n de verosimilitud negativa de GMM:
```python
from scipy.optimize import minimize

# ParÃ¡metros iniciales
x0 = parametros_iniciales

# LÃ­mites para los parÃ¡metros
bounds = [(None, None), (None, None), (1e-3, None), (1e-3, None), (0, 1)]

# MinimizaciÃ³n
resultado = minimize(gmm_verosimilitud_negativa, x0, method='L-BFGS-B', bounds=bounds)
```

En este ejemplo:
- `gmm_verosimilitud_negativa` es la funciÃ³n objetivo.
- `x0` son los valores iniciales de los parÃ¡metros.
- `method` es `'L-BFGS-B'`.
- `bounds` asegura que los parÃ¡metros como las desviaciones estÃ¡ndar sean positivas y que las proporciones estÃ©n entre 0 y 1.

````python
# Definimos los parÃ¡metros iniciales para la optimizaciÃ³n
# [mu1, mu2, sigma1, sigma2, pi_1]
x0 = [10, 20, 5, 5, 0.5]

# Establecemos los lÃ­mites para los parÃ¡metros:
# - mu1 y mu2 no tienen restricciones (pueden ser cualquier valor real)
# - sigma1 y sigma2 deben ser mayores que 0 (por eso el lÃ­mite inferior es 0.001)
# - pi_1 debe estar en el rango [0, 1] (proporciÃ³n vÃ¡lida para una mezcla)
bounds = [(None, None), (None, None), (1e-3, None), (1e-3, None), (0, 1)]

# Realizamos la minimizaciÃ³n de la funciÃ³n de verosimilitud negativa
# Usamos el mÃ©todo 'L-BFGS-B', que es adecuado para problemas con restricciones de caja
resultado = minimize(
  gmm_verosimilitud_negativa, 
  x0, 
  method='L-BFGS-B', 
  bounds=bounds
  )

# Mostramos el resultado de la optimizaciÃ³n
resultado
````
